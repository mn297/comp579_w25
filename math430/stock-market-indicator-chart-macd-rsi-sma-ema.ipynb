```python
# Missing Value Elimination

df5= df4.dropna()
df5
```

```python
df5.isnull().sum()
```

```python
df5["Open"].astype(int)
df5["High"].astype(int)
df5["Low"].astype(int)
df5["Close"].astype(int)
df5["Adj Close"].astype(int)
```

```python
# Noise Detection


Noisy_data=pd.DataFrame()
    
for i in df5.columns:
    if i == "Open":
        outlier=df5[(df5['Open']>5000) & (df5['Adj Close']<1000) ]       # According to Scatter plot
        Noisy_data = Noisy_data._append(outlier)
    if i == "Open":
        outlier=df5[(df5['Open']<1000) & (df5['Adj Close']>6000) ]       # According to Scatter plot
        Noisy_data = Noisy_data._append(outlier)
    if i == "High":
        outlier=df5[(df5['High']>6000) & (df5['Adj Close']<1000) ]       # According to Scatter plot
        Noisy_data = Noisy_data._append(outlier)
    if i == "High":
        outlier=df5[(df5['High']<1000) & (df5['Adj Close']>6000) ]       # According to Scatter plot
        Noisy_data = Noisy_data._append(outlier)
    if i == "Low":
        outlier=df5[(df5['Low']<1000) & (df5['Adj Close']>6000) ]        # According to Scatter plot
        Noisy_data = Noisy_data._append(outlier)
    if i == "Close":
        outlier=df5[(df5['Close']<1000) & (df5['Adj Close']>6000) ]      # According to Scatter plot
        Noisy_data = Noisy_data._append(outlier)
        
        
Noisy_data      
```

```python
# Noise Elimination

df6=df5.drop(Noisy_data.index.values)
df6
```

```python
df1008.ta.bbands(length=20, append=True)

# Plot Bollinger Bands
fig = go.Figure()

# Add candlestick trace
fig.add_trace(go.Candlestick(x=df1008['date'],
                open=df1008['open'],
                high=df1008['high'],
                low=df1008['low'],
                close=df1008['close'], name='Candlesticks'))

# Add Bollinger Bands traces
fig.add_trace(go.Scatter(x=df1008['date'], y=df1008['BBU_20_2.0'], mode='lines', line=dict(color='red'), name='Upper Bollinger Band'))
fig.add_trace(go.Scatter(x=df1008['date'], y=df1008['BBM_20_2.0'], mode='lines', line=dict(color='black'), name='Middle Bollinger Band'))
fig.add_trace(go.Scatter(x=df1008['date'], y=df1008['BBL_20_2.0'], mode='lines', line=dict(color='blue'), name='Lower Bollinger Band'))


fig.update_xaxes(rangeslider_visible=True, rangeselector=dict(
    buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")])))
fig.update_layout(title='Bollinger Bands Visualization With Candlesticks', xaxis_title='Date', yaxis_title='Price')

layout = go.Layout(
    plot_bgcolor='rgb(255,255,255)',
    # Font Families
    font_family='Arial',
    font_color='rgb(82,82,82)',
    font_size=12,
    xaxis=dict(
        rangeslider=dict(
            visible=False )))


# Update options and show plot
fig.update_layout(layout)
fig.show()
```
